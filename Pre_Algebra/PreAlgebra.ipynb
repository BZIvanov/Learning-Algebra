# Multiplication

Multiplication is repeated addition. For example $ 2 * 3 $ is the same as $ 2 + 2 + 2 $

## Properties

### Associative property

We can group numbers in any order without changing the product

$$ (2 * 3) * 4 = 3 * (2 * 4) $$

### Commutative property

$$ 4 * 5 = 5 * 4 $$

### Distributive property

Multiplication is distributive over addition/subtraction

$$ 3(2 + 4) = 3 * 2 + 3 * 4 $$

### Identity property of one

Any number multiplied by 1 remains unchanged

$$ 5 * 1 = 5 $$

### Property of zero

Any number multiplied by 0 result in 0

$$ 5 * 0 = 0 $$

## Rules

Example of how we could represent the multiplication

$$ 3 * 9 = 9 + 9 + 9 $$

# Division

## Rules

Zero divided by any nonzero number is always zero

$$ 0 \div 5 = 0 $$

$$ \frac{0}{5} = 0 $$

---

$$ 3 \div 0 = \text{undefined} $$

---

$$ 20 \div 5 = \frac{20}{5} $$

---

If we have 4 pieces of the whole 4, we have all pieces.

$$ \frac{4}{4} = 1 $$

---

$$ -\frac{1}{5} = \frac{-1}{5} = \frac{1}{-5} $$

---

Rule about how we can divide numbers. For example numbers ending with 0 are divisible by 10 and even numbers are divisible by 2.

- **by 2** - any number that ends in an even number
- **by 3** - any number whose digital root is 3, 6 or 9
- **by 5** - any number that ends in 5 or 0

For example: is 768 divisible by 3?

$$ 768 = 7 + 6 + 8 = 21 = 2 + 1 = 3 $$

768 is divisible by 3, $ 768 : 3 = 256 $

# Exponent

## Rules

### Exponent

Exponent/power is repeated addition.

$$ 3^5 = 3 * 3 * 3 * 3 * 3 $$

### Power of 1

Any number raised to the power of 1 is just itself

$$ 9^1 = 9 $$

### Squared number

_Squared_ number - number to the power of 2

$$ 3^2 $$

### Cubed number

_Cubed_ number - number to the power of 3

$$ 3^3 $$

### Paranthesis

If the minus sign is not in paranthesis, it is not included in the base, when multiplying the exponent.

$$ -2^2 = -4 $$

$$ (-2)^2 = 4 $$

### Transforming

$$ \sqrt{9} = 9^{\frac{1}{2}} = 9^{0.5} $$

### Quotient rule

$$ \frac{2^5}{2^2} = \frac{2*2*2*2*2}{2*2} = 2^3 $$

### Integer exponents

$$ 3^3 = 27 $$

$$ 3^2 = 9 $$

$$ 3^1 = 3 $$

$$ 3^0 = 1 $$

$$ 3^{-1} = \frac{1}{3} $$

$$ 3^{-2} = \frac{1}{9} $$

### Examples

- example 1

$$ 10^9 * 10^2 = 10^{11} $$

- example 2

$$ 10^6 : 10^4 = 10^{6-4} = 10^2 $$

- example 3

$$ 10^{-50} \div 10^{50} = 10^{-50 - 50} = 10^{-100} $$

# Factoring

## Whole numbers

### Prime numbers

A prime number is a whole/natural number larger than 1, that is only divisible by itself and 1.

2 is the only even prime number. Any even number larger than 2 is divisible by 2 and therefore not a prime number.

### Composite numbers

A composite number is a whole number that is greater than 1 and is divisible by a whole number other than itself and 1.

#### Examples

- example 1

185, is it prime or composite?

185 is composite, because it is divisible by 5.

- example 2

42, is it prime or composite?

42 is composite, because it is divisible by 6.

## GCF (Greates Common Factor)

GCF is the largest number that each number of a group is divisible by.

Greatest Common Divisor (GCD) and Greatest Common Factor (GCF) are the same thing.

### Examples

We need to create factor tree for each number. The factor tree for each number is reduced to prime numbers.

- example 1

$$ \text{GCF (36,54)} $$

$$ \text{branch 36} $$

$$ 36=6*6 $$

$$ 6=2*3 $$
$$ 6=2*3 $$

$$ \text{2 and 3 are prime numbers so we stop here} $$

$$ \text{branch 54} $$

$$ 54=9*6 $$

$$ 9=3*3 $$
$$ 6=2*3 $$

$$ \text{2 and 3 are prime numbers so we stop here} $$

$$ 36=2*2*3*3 $$
$$ 54=2*3*3*3 $$

$$ \text{the common numbers from both lists are 2, 3 and 3} $$

$$ \text{GCF(36,54) = 2 * 3 * 3 = 18} $$

- example 2

Find GCF for 20, 45 and 60.

$$ 20 = 4 * 5 = 2 * 2 * 5 $$
$$ 45 = 9 * 5 = 3 * 3 * 5 $$
$$ 60 = 6 * 10 = 2 * 3 * 2 * 5 $$

|  | | | | | |
|- |-|-|-|-|-|
|20|2|2|-|-|5|
|45|-|-|3|3|5|
|60|2|2|3|-|5|

$$ \text{We can see the only common for all is 5 so GCF = 5} $$

- example 3

Find GCF for 120, 168 and 216.

$$ 120 = 12 * 10 = 4 * 3 * 5 * 2 = 2 * 2 * 3 * 5 * 2 $$
$$ 168 = 42 * 4 = 7 * 6 * 2 * 2 = 7 * 3 * 2 * 2 * 2 $$
$$ 216 = 54 * 4 = 9 * 6 * 2 * 2 = 3 * 3 * 3 * 2 * 2 * 2 $$

|   | | | | | | | | |
|---|-|-|-|-|-|-|-|-|
|120|2|2|2|3|-|-|5|-|
|168|2|2|2|3|-|-|-|7|
|216|2|2|2|3|3|3|-|-|

$$ \text{We can see the common for all are 2,2,2,3 so GCF = 2*2*2*3 = 24} $$

## LCM (Least Common Multiple)

**4:** 4,8,12,16,20,24,28,32,36,40...

**5:** 5,10,15,20,25,30,35,40...

Common multiple for 4 and 5 are 20, 40 etc...

LCM(4,5) = 20

Finding LCM like the above example works for small numbers, because we can easyli list them, for bigger numbers is better to use factoring to find the prime numbers.

### Examples

- example 1

$$ \text{LCM(20,35)} $$

$$ \text{Apply factoring to find the prime numbers for each number} $$

$$ 20=2*2*5 $$
$$ 35=5*7 $$


|n  | p1| p2| p3|p4 |
|---|---|---|---|---|
|20 |2  |2  |5  |-  |
|35 |-  |-  |5  |7  |

$$ \text{LCM(20,35)}=2*2*5*7=140 $$

# Fractions

## Improper fraction to Mixed number

When we have improper fraction as solution, we can change it to mixed number.

### Examples

As we can see after the equal sign the number before the fraction is number of the whole pieces. For the numerator of the fraction the value is the remainder of the division and for the denominator of the fraction the value is the denominator of the fraction before the equal sign.

$$ \frac{18}{7}=2^{\frac{4}{7}} $$

$$ \frac{19}{3}=6^{\frac{1}{3}} $$

$$ \frac{11}{7}=1^{\frac{4}{7}} $$

$$ \frac{31}{15}=2^{\frac{1}{15}} $$

## Mixed number to Improper fraction

### Examples

$$ 3^{\frac{1}{8}}=\frac{3*8+1}{8}=\frac{25}{8} $$

$$ 8^{\frac{1}{3}}=\frac{8*3+1}{3}=\frac{25}{3} $$

$$ 3^{\frac{5}{8}}=\frac{3*8+5}{8}=\frac{29}{8} $$

$$ 100^{\frac{7}{9}}=\frac{100*9+7}{9}=\frac{907}{9} $$

## Multiplying fractions

Multiply the numerator from the first fraction by the numerator from the second fraction. Do the same for the denominators.

### Examples

- example 1

$$ \frac{2}{5} * \frac{1}{7} = \frac{2}{35} $$

- example 2

$$ \frac{3}{5} * \frac{10}{12} = \frac{3 * 5 * 2}{5 * 2 * 2 * 3} = \frac{1}{2} $$

- example 3

$$ \frac{9}{20}  *\frac{5}{33} = \frac{3 * 3}{5 * 2 * 2} * \frac{5}{3 * 11} = \frac{3}{2 * 2} * \frac{1}{11} = \frac{3}{44} $$

- example 4

$$ 6 * \frac{1}{3} = \frac{6}{1} * \frac{1}{3} = \frac{2}{1} = 2 $$

## Dividing fractions

It is easier to change the division to multiplication by multiplying to the reciprocal of the fraction.

### Examples

- example 2

$$ \frac{2}{9}:-4=\frac{2}{9}:-\frac{4}{1}=\frac{2}{9}*-\frac{1}{4}=\frac{1}{9}*-\frac{1}{2}=-\frac{1}{18} $$

- example 3

$$ \frac{3}{5}:\frac{21}{10}:\frac{9}{11}=\frac{3}{5}*\frac{10}{21}*\frac{11}{9}=\frac{1}{1}*\frac{2}{7}*\frac{11}{9}=\frac{22}{63} $$

- example 4

$$ \frac{8}{9} : \frac{3}{10} = \frac{8}{9} * \frac{10}{3} = \frac{80}{27} = 2\frac{26}{27} $$

## Adding and Subtractring fractions

We are reducing the numbers to prime numbers.

### Examples

#### With different denominators

For solving these tasks we need to find the LCD, which is LCM for the denominators.

- example 1

$$ \text{LCD->LCM(3,4)=12} $$

$$ \frac{1}{3}+\frac{1}{4}=\frac{1}{3}*\frac{4}{4}+\frac{1}{4}*\frac{3}{3}=\frac{4}{12}+\frac{3}{12}=\frac{4+3}{12}=\frac{7}{12} $$

- example 2

$$ \text{LCD->LCM(4,2)=4} $$

$$ \frac{1}{4}+\frac{1}{2}=\frac{1}{4}+\frac{1}{2}*\frac{2}{2}=\frac{1}{4}+\frac{2}{4}=\frac{3}{4} $$

- example 3

$$ \text{15=3*5} $$
$$ \text{25=5*5} $$
$$ \text{LCD->LCM(15,25)=3*5*5=75} $$

$$ \frac{14}{15}-\frac{17}{25}=\frac{14}{15}*\frac{5}{5}-\frac{17}{25}*\frac{3}{3}=\frac{70}{75}-\frac{51}{75}=\frac{19}{75} $$

- example 4

When adding fractions, we can also use cross multiply:

$$ \frac{3}{7} + \frac{4}{11} = \frac{(3 * 11) + (7 * 4)}{7 * 11} = \frac{33 + 28}{77} = \frac{61}{77} $$

### Equal denominators

The fraction with greater numerator is larger

- example 1

The fraction with smaller denominator is larger

- example 1

$$ \frac{3}{5} < \frac{3}{4}  $$

# Decimals

## Rules

Depending on how many digits we have after the decimal dot, we will add the same amount of zeros.

$$ 12.011 = 10 + 2 + \frac{0}{10} + \frac{1}{100} + \frac{1}{1000} $$

or in other words, we can also express like this:

$$ 12.011 = 12\frac{11}{1000} $$

## Decimal greater than 1 to fraction

When we have decimal, which is greater than 1, we will first convert it to mixed number and then to fraction

- example 1

$$ 13.7 = 13\frac{7}{10} = \frac{137}{10} $$

- example 2

$$ 5.25 = 5\frac{25}{100} = 5\frac{1}{4} = \frac{21}{4} $$

## Decimal less than 1 to fraction

When we have decimal, which is less than 1, we can directly convert it to fraction. We are not going to first convert to mixed number simply, because we don't have whole part/number.

Add as many zeros in the denominator as many decimal places are moved for the numerator.

- example 1

$$ 0.27 = \frac{0.27}{1} = \frac{27}{100} $$

- example 2

$$ 0.005 = \frac{0.005}{1} = \frac{5}{1000} $$
